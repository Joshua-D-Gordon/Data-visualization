```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

cats_str = r'C:\Users\Joshua Student\Desktop\temp\train'

current_df = pd.DataFrame(columns = list(range(0,784)))
size = 30
#print(current_df)

for i in range(0,12499,1):
    
    path = r'{}{}{}.jpg'.format(cats_str,'\cat.',i)
    
    img_gray = cv2.imread(path,0)
    
    resized_img = cv2.resize(img_gray, (size,size),interpolation = cv2.INTER_AREA)
    
    scaler = MinMaxScaler()

    df = pd.DataFrame(data=resized_img)

    df_norm = pd.DataFrame(scaler.fit_transform(resized_img))

    df_norm.to_numpy()

    df_norm = resized_img.reshape(size*size)


    df = pd.DataFrame(df_norm.transpose())

    final_value = df.transpose()     
    #print(final_value)
    final_value = final_value.divide(other = 255)
    
    #print(final_value)
    df_merged = pd.concat([current_df, final_value], axis=0)
    
    current_df = df_merged
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

X = reduced
y = df_merged['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 1)

clf_knn = KNeighborsClassifier(150)

clf_knn.fit(X_train, y_train)
score = clf_knn.score(X_test, y_test)
score
```
